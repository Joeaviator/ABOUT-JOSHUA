```python
import pandas as pd
import sqlite3
```

```python
database='database.sqlite'
conn=sqlite3.connect(database)
print("Database connected sucessfully")
```

```python
tables=pd.read_sql("""SELECT * FROM sqlite_master WHERE type='table';""",conn)
tables
```

```python
#match details
matches=pd.read_sql("""SELECT * FROM Match;""",conn)
matches
```

```python
matches.isnull().sum()
```

```python
#Drop
matches.dropna(inplace=True)
```

```python
#number of matches per season
results=pd.read_sql("""SELECT Season_Id, COUNT(*) AS TOTAL_MATCHES FROM Match GROUP BY Season_Id ORDER BY Season_Id;""",conn)
results
```

```python
team_result=pd.read_sql("""SELECT Match_Winner, COUNT(*) AS Wins FROM Match GROUP BY Match_Winner ORDER BY WINS DESC;""",conn)
team_result
```

```python
team=pd.read_sql("""SELECT * FROM Team;""",conn)
team
```

```python
#Average win margin per season
win_margin=pd.read_sql("""SELECT Season_Id, AVG(Win_Margin) AS Average_Win_Margin FROM Match GROUP BY Season_Id ORDER BY Average_Win_Margin DESC;""",conn)
win_margin
```

```python
#Number of matches played at each venue
venue_result=pd.read_sql("""SELECT Venue_Id, COUNT(*) AS Matches_Played FROM Match GROUP BY Venue_Id ORDER BY Matches_Played DESC""", conn)
venue_result
```

```python
#Count matches per winner per season
match_wins=pd.read_sql("""SELECT Season_Id, Match_Winner, COUNT(*) AS Wins
FROM Match
GROUP BY Season_Id,Match_Winner
ORDER BY Wins DESC;""", conn)
match_wins
```

```python
#Top 5 players with the man of the match awards
man_of_match=pd.read_sql("""SELECT Man_of_the_Match, COUNT(*) AS AWARDS FROM Match GROUP BY Man_of_the_Match ORDER BY AWARDS DESC LIMIT 5;""",conn)
man_of_match
```
